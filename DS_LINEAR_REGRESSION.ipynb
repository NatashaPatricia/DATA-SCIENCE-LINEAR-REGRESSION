DIVIDE THE DATA INTO TRAIN AND TEST SETS

```python
from sklearn.model_selection import train_test_split
x = pd.DataFrame(road['Age_of_Vehicle'])
y = pd.DataFrame(road['Number_of_Casualties'])
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

TRAIN THE ALGORITHM,,,RETRIEVE THE INTERCEPT,,,RETRIEVE THE SLOPE

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
print("Intercept is: ",regressor.intercept_)
print("Slope is: ",regressor.coef_)
```

PREDICT VALUES,,OUTPUT THE ACTUAL VALUES

```python
y_pred = regressor.predict(x_test)
print(y_pred)
y_test.head(7)
```

EVALUATE THE ALGORITHM

```python

from sklearn import metrics
import numpy as np
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Absolute Error", mae)
print("Mean Squared Error", mse)
print("Root Mean Squared Error", rmse)
```

SAVE THE LINEAR REGRESSION MODEL

```python
import joblib
import pickle
#train my linear regression model
model = LinearRegression()
model.fit(x,y)
#save the trained model to a file using joblib
joblib.dump(model, 'kasuku.pkl')
# Or Save the model using pickle
with open('kasuku.pkl', 'wb') as file:
  pickle.dump(model, file)
```
