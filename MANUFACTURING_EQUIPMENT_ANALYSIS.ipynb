```python
import pandas as pd
condition = pd.read_csv('/content/MANUFACTURING EQUIPMENT.csv')
condition.head(30)
```

CHECKING FOR DUPLICATES

```python
duplicates= condition[condition['Process temperature [K]'].duplicated()]
duplicatespeed = condition[condition['Rotational speed [rpm]'].duplicated()]
duplicatetool= condition[condition['Tool wear [min]'].duplicated()]
print(duplicates)
print(duplicatespeed)
print(duplicatetool)
```

DROPING DUPLICATE VALUES

```python
condition = condition.drop_duplicates()
df_cleaned = condition.drop_duplicates()
print(df_cleaned)
```

Print number of occurence

```python
unique_values = condition.value_counts()
print(unique_values)
condition.head(30)
```

FINDING NULL VALUES

```python
null = condition.isnull().sum()
print(null)
```

STATISTICAL SUMMARY REPORT

```python
report = condition.describe()
print(condition.info())
print(report)
```

VISUALIZING THE CHANGE IN VARIABLES

```python
import matplotlib.pyplot as plt
# Assuming X is your independent variable and y1 is one of your dependent variables
plt.scatter(condition['Process temperature [K]'], condition['Rotational speed [rpm]'], label='condition')
plt.xlabel('Tool wear [min]')
plt.ylabel('Rotational speed [rpm]')
#plt.ylabel('Process temperature [K] ')
plt.title('Scatter Plot of Data')
plt.legend()
plt.show()
```


ADDING A REGRESSION LINE

```python
from sklearn.linear_model import LinearRegression
X = condition[['Tool wear [min]']]
y = condition[['Rotational speed [rpm]']]
#y = condition[['Process temperature [K]']]
regressor = LinearRegression()
regressor.fit(X,y)
condition.plot(x='Tool wear [min]',y='Rotational speed [rpm]',style='*')
#condition.plot(x='Tool wear [min]',y='Process temperature [K]',style='*')
plt.plot(X, regressor.predict(X), color='cyan', label='Regression line')
plt.xlabel('Tool wear [min]')
plt.ylabel('Rotational speed [rpm]')
#plt.ylabel('Process temperature [K]')
plt.show()
```

DIVIDE THE DATA INTO TRAIN AND TEST SETS

```python
from sklearn.model_selection import train_test_split
x = pd.DataFrame(condition['Tool wear [min]'])
y = pd.DataFrame(condition['Rotational speed [rpm]'])
#y = pd.DataFrame(condition['Process temperature [K]'])
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```


TRAIN THE ALGORITHM,,,RETRIEVE THE INTERCEPT,,,RETRIEVE THE SLOPE

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
print("Intercept is: ",regressor.intercept_)
print("Slope is: ",regressor.coef_)
```

PREDICTION

```python
y_pred = regressor.predict(x_test)
print(y_pred)
y_test.head(7)
```

EVALUATE THE ALGORITHM

```python
from sklearn import metrics
import numpy as np
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Absolute Error", mae)
print("Mean Squared Error", mse)
print("Root Mean Squared Error", rmse)
```

SAVE THE LINEAR REGRESSION MODEL

```python
import joblib
import pickle
#train my linear regression model
model = LinearRegression()
model.fit(x,y)
#save the trained model to a file using joblib
joblib.dump(model, '.pkl')
# Or Save the model using pickle
with open('MANUFACTURING EQUIPMENT.pkl', 'wb') as file:
  pickle.dump(model, file)
```

USING THE TRAINED DATA TO TEST MORE DATA SET

```python
new_x = np.array([10,20,30,40,50]).reshape(-1,1)
model = joblib.load('MANUFACTURING EQUIPMENT.pkl')
predictions = model.predict(new_x)
new_data=pd.DataFrame({'x':new_x.flatten(), 'Predicted_y': predictions.flatten()})
print(new_data)
```
