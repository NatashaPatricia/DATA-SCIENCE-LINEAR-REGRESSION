VISUALIZING THE CHANGE IN VARIABLES

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
X = condition[['Process temperature [K]', 'Rotational speed [rpm]']]
y = condition['Tool wear [min]']
# Create a Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to your data
rf_model.fit(X, y)
# Get feature importances from the model
feature_importances = rf_model.feature_importances_
feature_names = X.columns

# Sort the features by importance
feature_indices = np.argsort(feature_importances)[::-1]

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_importances)), feature_importances[feature_indices], align="center")
plt.xticks(range(len(feature_importances)), feature_names[feature_indices], rotation=45)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Random Forest Feature Importance')
plt.show()
# Make predictions
predictions = rf_model.predict(X)

# Plot predictions vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(y, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Random Forest Predictions vs. Actual Values')
plt.show()


```


ADDING A REGRESSION LINE

DIVIDE THE DATA INTO TRAIN AND TEST SETS

```python
from sklearn.model_selection import train_test_split
x = pd.DataFrame(condition['Tool wear [min]'])
y = pd.DataFrame(condition['Rotational speed [rpm]'])
#y = pd.DataFrame(condition['Process temperature [K]'])
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```


TRAIN THE ALGORITHM,,,RETRIEVE THE INTERCEPT,,,RETRIEVE THE SLOPE

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
print("Intercept is: ",regressor.intercept_)
print("Slope is: ",regressor.coef_)
```

PREDICTION

```python
y_pred = regressor.predict(x_test)
print(y_pred)
y_test.head(7)
```

EVALUATE THE ALGORITHM

```python
from sklearn import metrics
import numpy as np
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Mean Absolute Error", mae)
print("Mean Squared Error", mse)
print("Root Mean Squared Error", rmse)
```

SAVE THE LINEAR REGRESSION MODEL

```python
import joblib
import pickle
#train my linear regression model
model = LinearRegression()
model.fit(x,y)
#save the trained model to a file using joblib
joblib.dump(model, '.pkl')
# Or Save the model using pickle
with open('MANUFACTURING EQUIPMENT.pkl', 'wb') as file:
  pickle.dump(model, file)
```
